```python
import pandas as pd
import numpy as np
```

```python
history__ = pd.read_pickle('Neural_network_history.pickle')
history__.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.plot(history__["loss"], label="Training Loss")
plt.plot(history__["val_loss"], label="Validation Loss")
plt.yscale("log", base=10)
plt.ylim(0.02,1.0)
#plt.yticks(np.arange(0.01, 0.01, 1.0))
plt.rc('grid', linestyle="-.", color='black')
plt.grid(True)
plt.text(250, history__['loss'][249], str(history__['loss'][249]), horizontalalignment='left', fontweight='bold', color='blue')
plt.text(250, history__['val_loss'][249], str(history__['val_loss'][249])+'0', horizontalalignment='left', fontweight='bold', color='blue')
plt.legend()
plt.xlabel("Epoch number", fontweight='bold')
plt.ylabel("Loss value", fontweight='bold')
plt.savefig('model_loss_plot_log.pdf', bbox_inches='tight')  
plt.show()
```
