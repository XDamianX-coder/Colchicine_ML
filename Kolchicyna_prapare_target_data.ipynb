## Convert to float 

```python
def comma_to_dot(float_num):
    
    return float_num.replace(',', '.')
```

```python
comma_to_dot(file['A549'][0])
```

```python
file['A549'][0]
```

```python
float(comma_to_dot(file['A549'][0]))*2
```

```python
def convert_target(num):
    
    try:
        new = float(comma_to_dot(num))
        #new = float(new)
    except:
        new = float(num)
        
    return new
```

```python
file.columns.to_list()[4:9]
```

```python
new_cols = file.columns.to_list()[4:9]
l_l = [str(name+'_') for name in new_cols]

for i, column in enumerate(new_cols):
    l_l[i] = [convert_target(x) for x in file[column]]
```

```python
for element in l_l: 
    for i in range(len(element)):

        res = str(str(i)+' '+str(element[i])+ ' '+str('Float? ')+str(isinstance(element[i], float)))

        if "True" in res:
            pass
        else:
            print(res)
    print("If everything is correct no (other) output will be produced...")
```

```python
for i, element in enumerate(file.columns.to_list()[4:9]):
    file[str(element+'_float')] = l_l[i]
```
